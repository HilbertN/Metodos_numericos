# Diagrama de Telara√±a

```python
import matplotlib.pyplot as plt
import numpy as np

def cobweb_tentmap(init,mu,nmax):
  t = np.zeros(nmax)
  t1 = np.zeros(nmax)
  t2 = np.zeros(nmax)
  t[0] = init
  halfm = nmax//2

  for n in range(1, nmax):
      if 0 <= t[n - 1] < 0.5:
          t[n] = float(mu * t[n - 1])
      elif 0.5<=t[n - 1] < 1:
          t[n] = float(mu * (1 - t[n - 1]))
      t[n] = round(t[n],20)

  for n in range(halfm):
      t1[2*n] = t[n]
      t1[2*n+1] = t[n]

  t2[0] = 0
  t2[1] = t[1]
  for n in range(1, halfm):
      t2[2*n] = t[n]
      t2[2*n+1] = t[n+1]
  print(t1)
  plt.figure()
  plt.plot(t1, t2, 'k')
  plt.plot([0, 0.5, 1], [0, mu/2, 0], 'k', linewidth=0.8)
  plt.plot([0, 1], [0, 1], 'k', linewidth=0.8)
  plt.xlabel('x')
  plt.ylabel('T')
  plt.axis([0, 1.2, 0, 1.2])
  plt.show()
```

```python
cobweb_tentmap(11/16,2,50)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def cobweb_Ex2(init,mu,nmax):
  t = np.zeros(nmax)
  t1 = np.zeros(nmax)
  t2 = np.zeros(nmax)
  t[0] = init
  halfm = nmax//2

  for n in range(1, nmax):
    t[n]=mu*t[n-1] *(1-t[n-1])



  for n in range(halfm):
      t1[2*n] = t[n]
      t1[2*n+1] = t[n]

  t2[0] = 0
  t2[1] = t[1]
  for n in range(1, halfm):
      t2[2*n] = t[n]
      t2[2*n+1] = t[n+1]
  x = np.linspace(0, 1, 100)
  y = mu*x*(1-x)


  plt.figure()
  plt.plot(t1, t2, color='black')
  plt.plot(x, y , color='black')
  plt.plot([0, 1], [0, 1], color='black', linewidth=0.8)
  plt.scatter([0.6875,0],[0.6875,0], color='blue',s=10)
  plt.axis([-0.01, 1, -0.01, 1])
  plt.xlabel('x')
  plt.ylabel('T')
  plt.savefig('ex2mu32pe2.png')
  plt.show()
```

```python
cobweb_Ex2(0.65,3.2,50)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def cobweb_Ex5(init,nmax):
  t = np.zeros(nmax)
  t1 = np.zeros(nmax)
  t2 = np.zeros(nmax)
  t[0] = init
  halfm = nmax//2

  for n in range(1, nmax):
    t[n]=(2*t[n-1])/(1+t[n-1])

  for n in range(halfm):
      t1[2*n] = t[n]
      t1[2*n+1] = t[n]

  t2[0] = 0
  t2[1] = t[1]
  for n in range(1, halfm):
      t2[2*n] = t[n]
      t2[2*n+1] = t[n+1]
  x = np.linspace(0, 2, 100)
  y = (2*x)/(1+x)


  plt.figure()
  plt.plot(t1, t2, color='black')
  plt.plot(x, y , color='black')
  plt.plot([0, 2], [0, 2], color='black', linewidth=0.8)
  plt.scatter([1],[1], color='blue',s=10)
  plt.axis([0, 2, 0, 2])
  plt.xlabel('x')
  plt.ylabel('T')
  plt.savefig('ex5pi2.png')
  plt.show()
```

```python
cobweb_Ex5(1.2,50)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

def cobweb_Ex6(init,nmax):
  t = np.zeros(nmax)
  t1 = np.zeros(nmax)
  t2 = np.zeros(nmax)
  t[0] = init
  halfm = nmax//2

  for n in range(1, nmax):
    t[n]=(1/2)*(t[n-1]+(3/t[n-1]))

  for n in range(halfm):
      t1[2*n] = t[n]
      t1[2*n+1] = t[n]

  t2[0] = 0
  t2[1] = t[1]
  for n in range(1, halfm):
      t2[2*n] = t[n]
      t2[2*n+1] = t[n+1]
  x = np.linspace(-5, 5, 100)
  y = (1/2)*(x+(3/x))


  plt.figure()
  plt.plot(t1, t2, color='black')
  plt.plot(x, y , color='black')
  plt.plot([-5, 5], [-5, 5], color='black', linewidth=0.8)
  plt.scatter([math.sqrt(3),-math.sqrt(3)],[math.sqrt(3),-math.sqrt(3)], color='blue',s=10)
  plt.axis([-5, 5, -5, 5])
  plt.plot([-5,5],[0,0],color='blue')
  plt.plot([0,0],[-5,5],color='blue')
  plt.xlabel('x')
  plt.ylabel('T')

  plt.savefig('ex6pi2.png')
  plt.show()
```
